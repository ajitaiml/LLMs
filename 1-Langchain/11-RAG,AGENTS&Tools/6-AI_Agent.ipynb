```python
# Modern 2026 Imports
from langgraph.prebuilt import create_react_agent
from langchain_openai import ChatOpenAI # Or your preferred LLM
from langchain_core.messages import HumanMessage

# 1. Initialize Model and Tools
llm = ChatOpenAI(model="gpt-4o")
tools = [search_tool, get_weather_data]

# 2. Create the Agent (Logic + Executor in one)
# This replaces Steps 2, 3, and 4 from your snippet.
agent = create_react_agent(llm, tools)

# 3. Invoke with a message list
query = "Find the Capital of Madhya Pradesh, then find its current weather condition"
result = agent.invoke({"messages": [HumanMessage(content=query)]})

# 4. Print the final response
print(result["messages"][-1].content)
```

```python
# Use the modern agent
from langgraph.prebuilt import create_react_agent

agent = create_react_agent(llm, tools)

inputs = {"messages": [("user", "Find the Capital of MP and its weather")]}

# Instead of .invoke(), we use .stream() to see the "behind the scenes"
for chunk in agent.stream(inputs, stream_mode="values"):
    message = chunk["messages"][-1]
    
    # This recreates the "Thought/Action" look
    if hasattr(message, "tool_calls") and message.tool_calls:
        print(f"--- Thought: I should use a tool ---")
        print(f"Action: {message.tool_calls[0]['name']}")
    elif message.type == "human":
        pass 
    else:
        print(f"Final Answer: {message.content}")
```

```python
# 1. Modern Imports (No more AgentExecutor or langgraph.prebuilt)
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

# 2. Initialize your LLM and Tools
llm = ChatOpenAI(model="gpt-4o")
tools = [search_tool, get_weather_data]

# 3. Create the Agent
# We set debug=True to see the internal "Thought/Action" logs
agent = create_agent(
    llm, 
    tools=tools, 
    debug=True, # This gives you the detailed trace you are looking for
    system_prompt="You are a helpful assistant that uses tools step-by-step."
)

# 4. Invoke (Step-by-step reasoning will now print to your console)
query = "Find the Capital of Madhya Pradesh, then find its current weather condition"
result = agent.invoke({"messages": [HumanMessage(content=query)]})

# 5. Print final result
print("\n--- FINAL ANSWER ---")
print(result["messages"][-1].content)
```
